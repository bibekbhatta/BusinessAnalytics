```python
import pulp
```

```python
model = pulp.LpProblem("Contribution_Margin", pulp.LpMaximize)
```

```python
# setting the number of units to be produced to integer and non-negative
Q = pulp.LpVariable('Q_var', lowBound=0, cat='Integer')
R = pulp.LpVariable('R_var', lowBound=0, cat='Integer')
S = pulp.LpVariable('S_var', lowBound=0, cat='Integer')
```

```python
# to optimize
model += (48 * Q) + (43 * R) + (39 * S)
```

```python
# constraints are as follows:
model += 12*Q + 10*R + 14*S <= 1200
model += 26*Q + 20*R + 17*S <= 2000
model += 45*Q + 40*R + 30*S <= 5000
model += 9*Q + 8*R + 8*S <= 1500
```

```python
# solving the porblem
model.solve()
```

```python
#presenting the status whether solution is found
print(pulp.LpStatus[model.status])
```

```python
#printing results
print ("Number of Q to be produced = {}".format(Q.varValue))
print ("Number of R to be produced = {}".format(R.varValue))
print ("Number of S to be produced = {}".format(S.varValue))
CM = pulp.value(model.objective)
print ('Total CM : ', CM)
```
